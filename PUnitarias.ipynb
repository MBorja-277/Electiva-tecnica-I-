```python
from datetime import date
class Persona:
  def __init__(self, n, e, fecha):
    self.__nombre = n
    self.Set_email(e)
    self.__años = self.calcularEdad(fecha)

  """
  el uso de "__" hace que los atributos se vuelvan privados
  """

  def calcularEdad(self, fecha):
    dias_del_año = 365.2425
    años = int((date.today() - fecha ).days / dias_del_año)
    #print(años)
    return años

  """
  Recibe como parámetro fecha (debe ser un objeto de tipo date, la fecha de nacimiento)
  Define cuántos días tiene un año en promedio
  Usa .2425 para considerar los años bisiestos

  date.today(): obtiene la fecha actual
  date.today() - fecha: resta la fecha de nacimiento de la fecha actual
  .days: convierte esa diferencia a días totales
  / dias_del_año: divide los días entre 365.2425 para obtener los años
  int(): convierte el resultado a número entero (eliminando así los decimales)
  depués retornará los años
  este metodo usa la libreria "from datetime import date"

  """

  def Set_email(self, e):
    if e.count("@") == 1 and e.find("@") >= 1 and e.find("@") + 2 < e.rfind("."):
      self.__email = e
      return e
    else:
      print("email incorrecto",e)
      self.__email = None

  """

Para obtener el email primero se deben verificar varias cosas para que el curreo cumpla:
el count se va a encargar de verificar que despues de un carácter haya un "@" y que debe existir ese "@".
despues  buscará la posición del "@" y le sumará 2 a esa posición para verificar que hayan por lo menos 2 carácteres entre el "@" y ultimo punto
y si todo esto cumple será guardado dentro e de lo contrario el correo será incorrecto y no tendra ningún valor

  """

  def ver_email(self):
    return self.__email

  def ver_nombre(self):
    return self.__nombre

  def set_cambiar_nombre(self, n):
    self.__nombre = n

  def get_cambiar_fecha(self):
    return self.__años

  """"
  Setter and getter
  """

  def __eq__(self, other):
    return self.__nombre == other.__nombre

  """"
  acá se busca hacer una comparación para eso es el "eq" se usará el self como el objeto actual y el other con el que se va a comparar
  """

  def __str__(self):
    return ("Su nombre es: " + self.__nombre + '\t email: ' + self.__email + '\t años: ' + str(self.__años))

  """"
  este dara el nombre va a dar un (espacio / tabulación) el correo (espacio / tabulación) junto con los años además convertirá los años a string
  """
```

```python
class Estudiante(Persona):

  def __init__(self, n, e, c, ID):
    super().__init__(n, e, c)  # n=nombre, e=email, c=fecha
    self.__SID = ID
    self.__Cursos = dict()
    self.__Promedio = 0

  """"
  el constructor, con el super se llama la clase persona
  y creamos un diccionario para los cursos de los estudiantes
  ademas de iniciar unas notas promedio en 0
  """

  def addCursos(self, CursoID, resultado):
    self.__Cursos[CursoID] = resultado
    self.calcularProm()

  """"
  acá se añaden cursos por medio de una lista y llama al método para calcular el promedio de las notas
  """

  def calcularProm(self):
    if len(self.__Cursos) > 0:
      total = sum(self.__Cursos.values())
      self.__Promedio = total / len(self.__Cursos)
    else:
      self.__Promedio = 0

  """"
  Si la longitud de los cursos es mayor a 0 el total será igual a la suma de todas las calificaciones que tiene el diccionario
  para eso es el values(). despuésel promedio será igual al resultado de la suma sobre la cantidad de cursos y de ahí se saca el promedio
  """

  def get_notasPromedio(self):
    return self.__Promedio

  def get_SID(self):
    return self.__SID

  def get_Cursos(self):
    return self.__Cursos

  def set_SID(self, ID):
    self.__SID = ID

  """"
  Setter and getter
  """

  def __str__(self):
    return (super().__str__() + '\t SID: ' + str(self.__SID) + '\t Cursos: ' + str(self.__Cursos) + '\t Promedio: ' + str(self.__Promedio))

  """"
  Pasa todos los datos a String para luego retornarlos
  """

  def __eq__(self, other):
    return isinstance(other,Estudiante) and self.__SID == other.__SID

  """"
  Verifica que other sea un objeto de tipo Student
  Previene errores si comparas con otro tipo de objeto
  Compara los IDs de ambos estudiantes
  Dos estudiantes son iguales si tienen el mismo ID
  """

  def __lt__(self, other):
    return self.__Promedio < other.__Promedio

  """"
  Compara por promedio de calificaciones
  Permite ordenar listas de estudiantes
  """
```
